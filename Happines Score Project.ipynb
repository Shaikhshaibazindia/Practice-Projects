```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
dfdata=pd.read_csv('https://raw.githubusercontent.com/dsrscientist/DSData/master/happiness_score_dataset.csv')
```

```python
dfdata.head()
```

```python
dfdata.tail()
```

```python
dfdata.info()
```

```python
dfdata.describe()
```

```python
#for checking for any null values
dfdata.isnull().sum()
```

Hence no null values are found in the data above

Now Plot between happiness score and GDP per capita

```python
plt.rcParams['figure.figsize']=(15,7)
plt.title('Plot between happiness score and GDP')
sns.scatterplot(x=dfdata['Happiness Score'], y=dfdata['Economy (GDP per Capita)'], hue=dfdata['Region'], s=200);
            
plt.legend(loc='upper left',fontsize='10')
plt.xlabel('Happiness Score')
plt.ylabel("Economy (GDP per Capita)")
```

In above scatter plot we can observe that GDP and Happiness score are directly proportional

```python
#Happiness score vs GDP
plot=plt.figure(figsize=(5,5))
sns.barplot(x='Happiness Score',y='Economy (GDP per Capita)',data=dfdata)
```

```python
# checking the overall GCP  region wise
gdp_region=dfdata.groupby('Region')['Economy (GDP per Capita)'].sum()
gdp_region
```

```python
gdp_region.plot.pie(autopct='%1.1f%%')
plt.title('GDP by region')
plt.ylabel('')
```

```python
# lets look for total countries in each region
total_country=dfdata.groupby('Region')[['Country']].count()
print(total_country)
```

```python
#corruption in the regions
corruption=dfdata.groupby('Region')[['Trust (Government Corruption)']].mean()
corruption
```

```python
plt.rcParams['figure.figsize']=(12,8)
plt.title('Perception of corruption in various Regions')
plt.xlabel('Regions',fontsize=15)
plt.ylabel('Corruption Index',fontsize=15)
plt.xticks(rotation=30,ha='right')
plt.bar(corruption.index, corruption['Trust (Government Corruption)'])
```

```python
top_10=dfdata.head(10)
bottom_10=dfdata.tail(10)
```

```python
fig,axes=plt.subplots(1,2,figsize=(16,6))
plt.tight_layout(pad=2)
xlabels=top_10.Country
axes[0].set_title("Top 10 happiest countries life Expectancy")
axes[0].set_xticklabels(xlabels,rotation=45,ha='right')
sns.barplot(x=top_10.Country,y=top_10['Health (Life Expectancy)'],ax=axes[0])
axes[0].set_xlabel('country Name')
axes[0].set_ylabel('life expectancy')

xlabels=bottom_10.Country
axes[1].set_title('Bottom 10 least happy countries life Expentancy')
axes[1].set_xticklabels(xlabels,rotation=45,ha='right')
sns.barplot(x=bottom_10.Country,y=bottom_10['Health (Life Expectancy)'],ax=axes[1])
axes[1].set_xlabel('Country Name')
axes[1].set_ylabel("Life expectancy")
```

```python
plt.rcParams['figure.figsize']=(15,7)
sns.scatterplot(x=dfdata.Freedom,y=dfdata['Happiness Score'],hue= dfdata.Region, s=200)
plt.legend(loc='upper left',fontsize='12')
plt.xlabel('Freedom of Choice')
plt.ylabel('Happiness Score')
```

```python
# lets see the top 10 countries with least corruption
country=dfdata.sort_values(by=['Trust (Government Corruption)']).head(10)
plt.rcParams['figure.figsize']=(12,6)
plt.title('Countries with most perception of corruption')
plt.xlabel('Country',fontsize=13)
plt.ylabel('corruption Index',fontsize=13)
plt.xticks(rotation=20, ha='right')
plt.bar(country.Country, country['Trust (Government Corruption)'])
```

```python
# lets see the top 10 countries with most corruption
country=dfdata.sort_values(by=['Trust (Government Corruption)']).tail(10)
plt.rcParams['figure.figsize']=(12,6)
plt.title('Countries with most perception of corruption')
plt.xlabel('Country',fontsize=13)
plt.ylabel('corruption Index',fontsize=13)
plt.xticks(rotation=20, ha='right')
plt.bar(country.Country, country['Trust (Government Corruption)'])
```
